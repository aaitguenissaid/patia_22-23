```python
import pandas as pd
import matplotlib.pyplot as plt

def graph(hsp_df, sat_df, image_name):
    # Replace zeros with NaN values
    hsp_df.replace(0.0, float('nan'))
    sat_df.replace(0.0, float('nan'))

    # Combine Makespan and Total_time columns from both DataFrames
    makespan_df = pd.concat([hsp_df['Makespan'], sat_df['Makespan']], axis=1)
    total_time_df = pd.concat([hsp_df['Total_time'], sat_df['Total_time']], axis=1)

    # Rename the columns
    makespan_df.columns = ['HSP', 'SAT']
    total_time_df.columns = ['HSP', 'SAT']


    # Sort the rows by the Makespan column
    makespan_df = makespan_df.sort_values('HSP')
    total_time_df = total_time_df.sort_values('HSP')

    # Extract Makespan and Total_time columns from each file
    hsp_makespan = makespan_df['HSP']
    sat_makespan = makespan_df['SAT']

    hsp_total_time = total_time_df['HSP']
    sat_total_time = total_time_df['SAT']

    # Create scatter plot comparing Makespan values
    fig1, ax1 = plt.subplots(figsize=(20, 10))
    ax1.plot(range(len(hsp_makespan)), hsp_makespan, label='HSP', marker='o')
    ax1.plot(range(len(sat_makespan)), sat_makespan, label='SAT', marker='o')
    ax1.set_xticks(range(len(hsp_makespan.index)))
    ax1.set_xticklabels(hsp_makespan.index)      # Set x-axis labels to index values
    ax1.set_xlabel('Problem Number')
    ax1.set_ylabel('Makespan')
    ax1.set_title('Makespan Comparison')
    ax1.legend()

    # Create scatter plot comparing Total_time values
    fig2, ax2 = plt.subplots(figsize=(20, 10))
    ax2.plot(range(len(hsp_total_time)), hsp_total_time, label='HSP', marker='o')
    ax2.plot(range(len(sat_total_time)), sat_total_time, label='SAT', marker='o')
    ax2.set_xticks(range(len(hsp_total_time.index)))
    ax2.set_xticklabels(hsp_total_time.index)      # Set x-axis labels to index values
    ax2.set_xlabel('Problem Number')
    ax2.set_ylabel('Total Time')
    ax2.set_title('Total Time Comparison')
    ax2.legend()
    
    fig1.savefig(image_name+'_makespan.png')
    fig2.savefig(image_name+'_total_time.png')
    
    plt.show()
```

```python
# Read in the first file
blocksworld_hsp_df = pd.read_csv('blocksworld_ipc2000_strips-typed_HSP_results.csv')

# Read in the second file
blocksworld_sat_df = pd.read_csv('blocksworld_ipc2000_strips-typed_results.csv')

graph(blocksworld_hsp_df, blocksworld_sat_df, 'blocksworld')
```
